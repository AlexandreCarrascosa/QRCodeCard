```python
from PIL import Image, ImageDraw, ImageFont
```

```python
font = ImageFont.truetype("Open_Sans/OpenSans-VariableFont_wdth,wght.ttf", size=12)

template = Image.open("images/template_card.png")
pic = Image.open("images/Foto1.jpg").resize((247, 300), Image.LANCZOS)
qrcode = Image.open("images/vcard_qr_Alexandre Carrascosa.png").resize((210, 210), Image.LANCZOS)

template.paste(pic, (20, 24))
template.paste(qrcode, (315, 100))

draw = ImageDraw.Draw(template)
draw.text((525,180), "Alexandre de F. Carrascosa", font=font, fill="black")
draw.text((525,220), "Tecnologia da Informao", font=font, fill="black")

template
```
